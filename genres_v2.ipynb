##Average of the attributes in each **genre**

```python
plt.figure(figsize = (6, 3), dpi=100)
plt.xticks(rotation = 90)
sns.countplot(df.index)
plt.show()
```

Underground Rap has the maximum number of instances while Pop has the minimum number of instances.

```python
plt.figure(figsize = (6, 3), dpi=100)
plt.xticks(rotation=90)
sns.barplot(data=df, x=df.index, y='danceability')
plt.show()
```

Songs in the genre Techhouse have the highest average danceability, followed by Underground Rap while Emo genre has the lowest average danceability

```python
plt.figure(figsize = (6, 3), dpi=100)
plt.xticks(rotation=90)
sns.barplot(data=df, x=df.index, y='loudness')
plt.show()
```

Songs in all the genres have loudness scores in negative. Songs in the genre Trap have the highest average loudness, followed by DNB while Techno genre has the lowest average loudness. 

```python
plt.figure(figsize = (6, 3), dpi=100)
plt.xticks(rotation=90)
sns.barplot(data=df, x=df.index, y='speechiness')
plt.show()
```

Songs in the genre Underground Rap have the highest average speechiness, followed by Rap while Psytrance genre has the lowest average speechiness.

```python
plt.figure(figsize = (6, 3), dpi=100)
plt.xticks(rotation=90)
sns.barplot(data=df, x=df.index, y='energy')
plt.show()
```

Songs in the genre Trap have the highest average energy, followed by Psytrance while RNB genre has the lowest average energy.

```python
plt.figure(figsize = (6, 3), dpi=100)
plt.xticks(rotation=90)
sns.barplot(data=df, x=df.index, y='acousticness')
plt.show()
```

Songs in the genre RNB have the highest average acousticness, followed by Hiphop while Psytrance genre has the lowest average acousticness

```python
plt.figure(figsize = (6, 3), dpi=100)
plt.xticks(rotation=90)
sns.barplot(data=df, x=df.index, y='instrumentalness')
plt.show()
```

Songs in the genre Techno have the highest average instrumentalness, followed by Psytrance while Rap genre has the lowest average instrumentalness.

```python
plt.figure(figsize = (6, 3), dpi=100)
plt.xticks(rotation=90)
sns.barplot(data=df, x=df.index, y='liveness')
plt.show()
```

Songs in the genre Trance have the highest average liveness, followed by Psytrance while Techhouse genre has the lowest average liveness.

```python
plt.figure(figsize = (6, 3), dpi=100)
plt.xticks(rotation=90)
sns.barplot(data=df, x=df.index, y='valence')
plt.show()
```

Songs in the genre Techhouse have the highest average valence, followed by Pop while Techno genre has the lowest average valence.

```python
plt.figure(figsize = (6, 3), dpi=100)
plt.xticks(rotation=90)
sns.barplot(data=df, x=df.index, y='tempo')
plt.show()
```

Songs in the genre DNB have the highest average tempo, followed by Hiphop while Techhouse genre has the lowest average tempo.

```python
plt.figure(figsize = (6, 3), dpi=100)
plt.xticks(rotation=90)
sns.barplot(data=df, x=df.index, y='duration_ms')
plt.show()
```

Songs in the genre Psytrance & Techno are longer than all other genres

```python
data = (df
        .groupby(['genre', 'mode'])['mode']
        .count()
        .unstack(1))

data.style.background_gradient(cmap=plt.cm.coolwarm_r)
```

Songs in the genre Underground Rap have the highest occurences in both modes.

```python
corr_mat = df.corr()
plt.figure(figsize = (9,9), dpi=100)
sns.heatmap(corr_mat,
            annot=True,
            fmt='.2f',
            cmap=plt.cm.coolwarm_r,
           )
plt.show()
```

Here positive value means that the two attributes have positive correlation between them.
